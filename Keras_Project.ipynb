**Import Required Packages.**

```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
```

**Read & Analyze Data.**

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
concrete_data.shape
concrete_data.isnull().sum()
concrete_data_columns = concrete_data.columns
```

**Create Input DataFrame with Predictors and Normalized Predictors.**

```python
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
predictors.head()
target.head()
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
n_cols = predictors_norm.shape[1] # number of predictors
```

**Define Model with 10 nodes and Parameters(optimizer= adam, loss metric = mean squared error and activation = relu).**

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    #Hidden layer of 10 nodes
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    #Output Layer
    model.add(Dense(1))
# compile model
    model.compile(loss='mse', optimizer='adam', metrics=['mse'])
    return model
# build the model
model = regression_model()
```

**PART-A : Fit Model & Perform 50 Iterations of 50 Epochs each with Non-Normalized Data. Also Display and Plot the Mean and SD of 50 sets of Mean Squared Errors.**

```python
# fit the model
iterations = pd.DataFrame()
# run 50 iterations
for i in range(50):
 history= model.fit(predictors, target, validation_split=0.3, epochs=50, verbose = 0) 
 err = pd.DataFrame(history.history['mean_squared_error'])
 err.columns=[i]
 iterations = pd.concat([iterations, err], axis=1)
view=iterations.describe().T
display(view)
plt.plot(view['std'])
plt.xlabel('Iteration Number->')
plt.ylabel('Standard Deviation of MSE')
plt.show()
plt.plot(view['mean'])
plt.xlabel('Iteration Number->')
plt.ylabel('Mean of MSE')
plt.show()
```

**We Note that both Mean and STD of losses tend to zero values with the mean stabilizing as the number of iterations increases.**

**PART-B : Fit Model & Perform 50 Iterations of 50 Epochs each with Normalized Data. Also Display and Plot the Mean and SD of 50 sets of Root Mean Squared Errors.**

```python
# fit the model
iterations = pd.DataFrame()
# run 50 iterations
for i in range(50):
 history= model.fit(predictors_norm, target, validation_split=0.3, epochs=50, verbose = 0) 
 err = pd.DataFrame(history.history['mean_squared_error'])
 err.columns=[i]
 iterations = pd.concat([iterations, err], axis=1)
view=iterations.describe().T
display(view)
plt.plot(view['std'])
plt.xlabel('Iteration Number->')
plt.ylabel('Standard Deviation of MSE')
plt.show()
plt.plot(view['mean'])
plt.xlabel('Iteration Number->')
plt.ylabel('Mean of MSE')
plt.show()
```

**We note that both Mean and STD of losses tend to zero values with the mean stabilizing as the number of iterations increases. Compared to the Non-Normalized data, the losses are much smaller to start with.**

**PART-C : Repeat Part B with 100 epochs.**

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    #model.add(Dense(50, activation='relu'))
    model.add(Dense(1))
# compile model
    model.compile(loss='mse', optimizer='adam', metrics=['mse'])
    return model
# build the model
model = regression_model()
# fit the model
iterations = pd.DataFrame()
# run 50 iterations
for i in range(50):
 history= model.fit(predictors_norm, target, validation_split=0.3, epochs=100, verbose = 0) 
 err = pd.DataFrame(history.history['mean_squared_error'])
 err.columns=[i]
 iterations = pd.concat([iterations, err], axis=1)
view=iterations.describe().T
display(view)
plt.plot(view['std'])
plt.xlabel('Iteration Number->')
plt.ylabel('Standard Deviation of MSE')
plt.show()
plt.plot(view['mean'])
plt.xlabel('Iteration Number->')
plt.ylabel('Mean of MSE')
plt.show()
```

**We see that the results between Parts B and C are similar with the losses tending much closer to 0 with increased iterations in Part C, compared to B.**

**PART-D : Repeat Part B with 3 hidden layers of 10 nodes each and Relu activation function.**

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    #First Hidden Layer
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    #Second Hidden Layer
    model.add(Dense(10, activation='relu'))
    #Third Hidden Layer
    model.add(Dense(10, activation='relu'))
    #Output Layer
    model.add(Dense(1))
# compile model
    model.compile(loss='mse', optimizer='adam', metrics=['mse'])
    return model
# build the model
model = regression_model()
# fit the model
iterations = pd.DataFrame()
# run 50 iterations
for i in range(50):
 history= model.fit(predictors_norm, target, validation_split=0.3, epochs=50, verbose = 0) 
 err = pd.DataFrame(history.history['mean_squared_error'])
 err.columns=[i]
 iterations = pd.concat([iterations, err], axis=1)
view=iterations.describe().T
display(view)
plt.plot(view['std'])
plt.xlabel('Iteration Number->')
plt.ylabel('Standard Deviation of MSE')
plt.show()
plt.plot(view['mean'])
plt.xlabel('Iteration Number->')
plt.ylabel('Mean of MSE')
plt.show()
```

**We can see that by increasing the number of layers, we now see a smaller mean but a larger STD, implying that the convergence of losses to zero is better but are more dispersed**
